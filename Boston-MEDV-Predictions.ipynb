## Question 1a

```python
from sklearn.linear_model import LinearRegression # Imports function used to build linear model
from sklearn.model_selection import train_test_split # Imports function to split data into train and test set
```

```python
# Defining X (features) and y (target) values 
X = df['RM']
y = df['MEDV']
```

```python
# Split the data into train and validation sets where the training set is 70% of the data - train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7, random_state=42)
```

```python
X_train.head() # Data is shuffled - can see that obs aren't in order anymore
```

```python
len(X_train) # See length of obs in that variable
```

```python
X_train_reshape = X_train.values.reshape(-1, 1) # Reshaping data based on "ValueError: Expected 2D array, got 1D array instead"
X_test_reshape = X_test.values.reshape(-1, 1)
```

```python
# Loading linear model
lm = LinearRegression()
lm.fit(X_train_reshape, y_train)
```

```python
print(f'Beta coefficient: {lm.coef_}') # Returns the coefficient, Beta, for X
print(f'Intercept: {lm.intercept_}') # Returns the constant, alpha
```

```python
# Plot the regression 
plt.figure(figsize=(10,5)) # Resize graph, length by height
plt.scatter(X_train_reshape, y_train) # Scatterplot
plt.plot(X_train_reshape, lm.intercept_ + lm.coef_*X_train_reshape, color="red") # Fitted line through graph
plt.xlabel('RM', fontsize=15)    # Label the x-axis
plt.ylabel('MEDV', fontsize=15)  # Label the y-axis
plt.title('Linear Regression of Median House Values on Average Number of Rooms per House', fontsize=15)
plt.show() # Gets rid of unnecessary text
```

```python
# Import mean squared error function because default scoring metric in model (LinearRegression()) is R-squared - we want the MSE
from sklearn.metrics import mean_squared_error 
```

```python
y_train_predict = lm.predict(X_train_reshape) # Using the coefficients from our lm model, predict the median house value (y_hat)
y_train_predict
```

```python
# Calculate the MSE of our training data 
train_linear_mse = mean_squared_error(y_train, y_train_predict)
print(f'MSE of linear model: {train_linear_mse}')
```

```python
# Predict the y_hats using the linear model we just found using the validation/test data set
y_test_predict = lm.predict(X_test_reshape) # Using the coefficients from our lm model, predict the median house value (y_hat)
y_test_predict
```

```python
# Calculate MSE of the validation set
test_linear_mse = mean_squared_error(y_test, y_test_predict)
print(f'MSE of test set: {test_linear_mse}')
```

## Polynomial Model

```python
# Preprocessing to create polynomial features for polynomial models
from sklearn.preprocessing import PolynomialFeatures
polynomial_features = PolynomialFeatures(degree=2) # Second order polynomial
```

```python
# Fit X_train and transform it
X_train_poly = polynomial_features.fit_transform(X_train_reshape)
# Transform the X_test too, but don't fit it 
X_test_poly = polynomial_features.transform(X_test_reshape)
```

```python
# Build a polynomial regression
pm = LinearRegression()
pm.fit(X_train_poly, y_train)
```

```python
# Show coefficient estimates for our polynomial model
print(f'Coefficients: {pm.coef_}')
print(f'Intercept: {pm.intercept_}')
```

```python
# Predict MEDV values using our pm model
y_train_predict_pm = pm.predict(X_train_poly) # Using the coefficients from our pm model, predict the median house value (y_hat)
y_train_predict_pm
```

```python
# Graph
plt.scatter(X_train_reshape, y_train)
plt.scatter(X_train_reshape, y_train_predict_pm)
```

```python
# Calculate the MSE of the polynomial model
train_polynomial_mse = mean_squared_error(y_train, y_train_predict_pm)
print(f'MSE of polynomial model: {train_polynomial_mse}')
```

```python
# Predict the y_hats using the pm we just found using the validation/test data set
y_test_predict_pm = pm.predict(X_test_poly) # Using the coefficients from our pm model, predict the median house value (y_hat)
y_test_predict_pm
```

```python
# Calculate the MSE of the polynomial model using test set
test_polynomial_mse = mean_squared_error(y_test, y_test_predict_pm)
print(f'MSE of polynomial model: {test_polynomial_mse}')
```

## For Loop for higher degree polynomials

```python
# Repeats the above process (from x-variable transformation to MSE estimation) for polynomial degrees 1-5

for i in range(1,11):
    # Specify degree of polynomial
    polynomial_features = PolynomialFeatures(degree=i)
    # Print the degree of flexibility
    print(f'Degree of flexibility: {i}')
    # Transform
    X_train_poly = polynomial_features.fit_transform(X_train_reshape) # Fit X_train and transform it   
    X_test_poly = polynomial_features.transform(X_test_reshape) # Transform the X_test too, but don't fit it 
    # Estimate the model
    pm = LinearRegression()
    pm.fit(X_train_poly, y_train)
    # Return coefficients
    print(f'Coefficients: {pm.coef_}')
    print(f'Intercept: {pm.intercept_}')
    # Predict y_hat with model
    y_train_predict_pm = pm.predict(X_train_poly) 
    y_train_predict_pm
    # Graph
    plt.scatter(X_train_reshape, y_train, label="training data")
    plt.scatter(X_train_reshape, y_train_predict_pm, label='model')
    plt.title(f'Flexibility {i}')
    plt.xlabel(f'RM')
    plt.ylabel(f'MEDV')
    plt.legend()
    plt.show()
    # Calculate the MSE of the polynomial model
    train_polynomial_mse = mean_squared_error(y_train, y_train_predict_pm)
    print(f'MSE of training set: {train_polynomial_mse}')
    # Predict the y_hats using test data set
    y_test_predict_pm = pm.predict(X_test_poly) # Using the coefficients from our pm model, predict the median house value (y_hat)
    y_test_predict_pm
    # Calculate the MSE of the model using test set
    test_polynomial_mse = mean_squared_error(y_test, y_test_predict_pm)
    print(f'MSE of test set: {test_polynomial_mse} \n \n \n')
```

```python
# Creating empty lists to store MSEs of training and test sets, as well as the degree of flexibility
MSE_train = []
MSE_test = []
Deg_Flexibility = []

# Repeats the above process (from x-variable transformation to MSE estimation) for polynomial degrees 1-5

for i in range(1,11):
    
    # Specify degree of polynomial
    polynomial_features = PolynomialFeatures(degree=i)
    
    # Print the degree of flexibility
    print(f'Degree of flexibility: {i}')
    
    # Transform
    X_train_poly = polynomial_features.fit_transform(X_train_reshape) # Fit X_train and transform it   
    X_test_poly = polynomial_features.transform(X_test_reshape) # Transform the X_test too, but don't fit it 
    
    # Estimate the model
    pm = LinearRegression()
    pm.fit(X_train_poly, y_train)
    
    # Return coefficients
    print(f'Coefficients: {pm.coef_}')
    print(f'Intercept: {pm.intercept_}')
    
    # Predict y_hat with model
    y_train_predict_pm = pm.predict(X_train_poly) 
    y_train_predict_pm
    
    # Graph
    plt.scatter(X_train_reshape, y_train, label="training data")
    plt.scatter(X_train_reshape, y_train_predict_pm, label='model')
    plt.title(f'Flexibility {i}')
    plt.xlabel(f'RM')
    plt.ylabel(f'MEDV')
    plt.legend()
    plt.show()
    
    # Calculate the MSE of the polynomial model
    train_polynomial_mse = mean_squared_error(y_train, y_train_predict_pm)
    print(f'MSE of training set: {train_polynomial_mse}')
    
    # Predict the y_hats using test data set
    y_test_predict_pm = pm.predict(X_test_poly) # Using the coefficients from our pm model, predict the median house value (y_hat)
    y_test_predict_pm
    
    # Calculate the MSE of the model using test set
    test_polynomial_mse = mean_squared_error(y_test, y_test_predict_pm)
    print(f'MSE of test set: {test_polynomial_mse} \n \n \n')
    
    # Add the MSEs and Flexibility values to a list
    MSE_train.append(train_polynomial_mse) # Add the MSEs of the training set to the list
    MSE_test.append(test_polynomial_mse) # Add the MSEs of the test set to the list
    Deg_Flexibility.append(i) # Add the degree of flexibility to the list
```

```python
# Display the MSEs and degree of flexibilty
print(f'MSE of training set: {MSE_train}')
print(f'MSE of test set: {MSE_test}')
print(f'Degree of Felixbility: {Deg_Felixibility}')
```

## Question 1c - Estimation using LOOCV

```python
# Import library
from sklearn.model_selection import cross_val_score, LeaveOneOut, KFold
```

```python
# Defining CV methods
loocv = LeaveOneOut() # Assigning LOOCV function to variable loocv
MSE_RM = []  # Create a list to store MSE values
Deg_Flexibility = []  # Create a list to store polynomial order

# Using For Loops to calculate  
for i in range(1,11):
    # Specify degree of polynomial
    polynomial_features = PolynomialFeatures(degree=i)
   
    # Print the degree of flexibility
    print(f'Degree of flexibility: {i}')
    
    # Transform
    RM_train = polynomial_features.fit_transform(X.values.reshape(-1,1)) # Calculate x^2, x^3, ..., x^i 
    
    # Estimate the model
    pm = LinearRegression()
    
    # Calculate LOOCV MSEs 
    loocv_mse = cross_val_score(pm, RM_train, y, scoring='neg_mean_squared_error', cv=loocv, n_jobs=-1)
    # Can choose to see what n-1 is using:  print(f'n-1 = {len(loocv_mse)}')
    
    # Add the MSEs and Flexibility values to a list
    MSE_RM.append(np.mean(abs(loocv_mse))) # Take the absolute value of the negative MSEs
    Deg_Flexibility.append(i) # Add the degree of flexibility to the list
    
    # Display the mean MSE of the model using RM to predict MEDV
    print(f'Mean MSE of model: {np.mean(abs(loocv_mse))} \n')
    
    
```

## Question 2a. 

```python
# Import Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
```

```python
# Defining variables, subset data to variables of interest 
X = df.loc[:, ['RM', 'AGE', 'LSTAT']] # loc[row index, column names] 
y = df['MEDV']

# Split into training and validation sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
# Construct the Regression Tree Model, split into regions
rt = DecisionTreeRegressor()
rt.fit(X_train, y_train)

# Display the tree
from sklearn import tree # Import Tree function
plt.figure(figsize=(40,40))
tree.plot_tree(rt, feature_names=X_train.columns, filled=True, fontsize=12)

# See how many leaves
print(f'No. of leaves: {rt.get_n_leaves()}')
```

```python
# Predict y_hat for the training and validation set
yhat_train = rt.predict(X_train)
yhat_test = rt.predict(X_test)
```

```python
# Calculate the MSE 
print(f'MSE of the training set: {mean_squared_error(y_train, yhat_train)}')
print(f'MSE of the test set: {mean_squared_error(y_test, yhat_test)}')
```

## Question 2b.

```python
# Use k-fold cross validation to examine how MSEs vary as a function of the complexity penalty, alpha: 
MSE_rt = []  # Create a list to store MSE values
alpha = []  # Create a list to store levels of alpha

# Using For Loops to calculate MSEs
for i in range(0,21): 
    # Print the degree of flexibility
    print(f'Alpha: {i}')
    
    # Construct the regression tree model, split into regions, using 100% of the data
    rt = DecisionTreeRegressor(ccp_alpha=i)
       
    # Calculate k Fold MSEs where k=10 folds
    # Note: cv = 10 means split data into 10 folds
    # Note: n_jobs=-1 means run using full computing power
    kfold_mse_rt = cross_val_score(rt, X, y, scoring='neg_mean_squared_error', cv=10, n_jobs=-1) 
    
    # Add the MSEs and Flexibility values to a list
    # Takes the absolute values of the negative MSEs, then take the mean of it, and add it to the MSE_rt list
    MSE_rt.append(np.mean(abs(kfold_mse_rt))) 
    alpha.append(i) # Add the levels of alpha to the alpha list
    
    # Display the mean MSE of the model using RM to predict MEDV
    print(f'Mean MSE of regression tree model using 10-fold: {np.mean(abs(kfold_mse_rt))} \n')
```

```python
# Plot a graph of MSEs vs 
plt.figure(figsize=(10,8)) # How big you want your graph to look wxh
plt.scatter(alpha, MSE_rt)
plt.xticks(range(0,21))
plt.title('Average MSE as a Function of Complexity Penalty Alpha Using k-Fold')
plt.xlabel('Alpha')
plt.ylabel('Average MSE')
plt.show()
```
